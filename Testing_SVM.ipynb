```python
import cv2
import numpy as np
from skimage.feature import hog
import joblib
```

```python
IMG_SIZE = (128, 128)
CATEGORIES = ['bacterial_spot', 'healthy']

hog_params = {
    'orientations': 9,
    'pixels_per_cell': (8, 8),
    'cells_per_block': (2, 2),
    'block_norm': 'L2-Hys'
}
```

```python
def extract_hog_features(image):
    image = cv2.resize(image, IMG_SIZE)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    features = hog(gray, **hog_params)
    return features
```

```python
# 1. Load image
image = cv2.imread('dataset/healthy/002f87b7-e1a5-49e5-a422-bb423630ded5___JR_HL 8068.JPG')
if image is None:
    raise ValueError("Image could not be loaded.")

# 2. Extract features
features = extract_hog_features(image).reshape(1, -1)

# 3. Load model
model = joblib.load('bell_pepper_svm_model.joblib')

# 4. Predict
prediction = model.predict(features)[0]
probability = model.predict_proba(features)[0][prediction]

print(f"Prediction: {CATEGORIES[prediction]}")
print(f"Confidence: {probability * 100:.2f}%")
```
